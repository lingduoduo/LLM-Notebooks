```python
from vllm import LLM, SamplingParams
import sys
import os

# Store original fileno methods
original_stdout_fileno = None
original_stderr_fileno = None

# Monkey-patch fileno for sys.stdout and sys.stderr if they exist
# This is a common workaround for vLLM in environments like Colab
# where sys.stdout/sys.stderr might not have a working fileno() method.
if hasattr(sys.stdout, 'fileno'):
    original_stdout_fileno = sys.stdout.fileno
    sys.stdout.fileno = lambda: 1 # Return a dummy file descriptor

if hasattr(sys.stderr, 'fileno'):
    original_stderr_fileno = sys.stderr.fileno
    sys.stderr.fileno = lambda: 2 # Return a dummy file descriptor

try:
    llm = LLM(
        model="facebook/opt-125m",
        tensor_parallel_size=1,
        enforce_eager=False,
        disable_log_stats=True,
    )
finally:
    # Restore original fileno methods
    if original_stdout_fileno:
        sys.stdout.fileno = original_stdout_fileno
    if original_stderr_fileno:
        sys.stderr.fileno = original_stderr_fileno

outputs = llm.generate(
    ["Hello, my name is"],
    SamplingParams(max_tokens=20)
)

print(outputs[0].outputs[0].text)
```

```python
from vllm import SamplingParams

sampling_params = SamplingParams(temperature=0.7, top_p=0.95, max_tokens=20)

prompts = [
    "Hello, my name is",
    "The president of the United States is",
    "The capital of France is",
    "The future of AI is"
]

outputs = llm.generate(prompts, sampling_params)

for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")
```
