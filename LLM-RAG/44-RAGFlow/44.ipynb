## Dify 联合 RAGFlow：全面升级语义理解与动态召回能力


## 一 优势互补

### Dify 的核心定位

作为一款面向开发者和业务人员的低代码 AI 应用开发平台，Dify 在易用性、灵活性与流程编排能力方面表现突出。

它支持可视化工作流设计，内置 Agent 模块，能够轻松实现条件判断、多步骤调用、外部工具集成等复杂逻辑。

更重要的是，Dify 原生支持跨知识库检索机制 —— 当用户提出问题时，系统可根据语义自动选择最匹配的知识源进行查询，显著提升了召回准确率和使用效率。

然而，在实际应用中我们也发现，一个智能问答系统的最终效果，不仅取决于上层的调度逻辑，更关键的是底层知识处理的质量。而这正是 RAGFlow 的强项。

### RAGFlow 的核心优势

RAGFlow 专注于非结构化文档的深度解析与结构化提取。

无论是多栏排版的 PDF、扫描图像，还是包含复杂表格的技术文档，它都能通过其自主研发的 DeepDoc 技术实现高精度的内容还原。

同时，RAGFlow 内置 OCR 能力，提供多种文本切分策略，并支持主流大模型接口接入，确保在不同场景下都能输出高质量的向量化数据。


### 怎样结合 Dify 与 RAGFlow

那么，当 Dify 遇上 RAGFlow，会产生怎样的化学反应？

我们可以这样理解：Dify 是整个系统的“大脑”与“指挥中心”，负责任务调度、流程控制与用户体验；

而 RAGFlow 则是系统的“眼睛”与“记忆中枢”，专注于从原始文档中提炼出真正可用的知识内容。

通过标准 API 接口，我们将 RAGFlow 的知识库服务能力嵌入到 Dify 的工作流之中。

这样一来，既保留了 Dify 强大的应用构建能力，又弥补了其在复杂文档处理方面的短板；同时，也让 RAGFlow 的专业解析能力得以在一个完整的交互系统中发挥作用。

这种架构的优势在于：

实现了功能解耦与职责分离，便于维护与扩展；
提升了整体系统的语义理解深度与动态召回精度；
支持未来灵活替换模型后端或多知识源聚合，具备良好的可演进性。

## 二、安装部署流程

### 步骤 1：获取 RAGFlow 源码

从官方 GitHub 仓库下载指定稳定版本（建议使用 v0.20.5）：

```bash
git clone -b v0.20.5 https://github.com/infiniflow/ragflow.git
```

### 步骤 2：修改端口避免冲突

由于 Dify 默认占用 `80` 和 `443` 端口，若在同一主机运行 RAGFlow，需调整其 `docker-compose.yml` 文件中的端口映射：

```yaml
ports:
  - "8080:80"   # 原为 80:80
  - "8443:443"  # 原为 443:443
```

此举可有效避免服务端口冲突。

### 步骤 3：启动 RAGFlow 服务

使用 Docker Compose 启动容器，并通过 `-p` 参数命名项目，便于区分不同服务：

```bash
docker compose -p ragflow up -d
```

等待所有服务就绪后，访问 `http://localhost:8080` 即可进入 RAGFlow 管理后台。


## 三、RAGFlow 配置步骤

### 1. 创建用户账户

首次登录时注册管理员账号，用于后续权限管理与知识库操作。

### 2. 配置模型供应商

进入「设置」→「模型供应商」页面，添加所需的大语言模型（LLM）接入信息，例如 OpenAI、通义千问、百川、Moonshot 等，确保推理服务正常可用。

### 3. 创建并初始化知识库

- 新建知识库，命名并选择适用领域；
- 上传测试文档（推荐包含 PDF、扫描图像、Excel 表格等典型复杂文件）；
- 观察 DeepDoc 是否成功完成结构化解析，检查段落划分与表格还原效果。

### 4. 启用 API 接口

前往「开发者中心」或「API 设置」页面，开启 API 访问权限，生成访问令牌（API Key），并记录核心接口地址，例如：

```
POST http://<your-host>:8080/v1/document/questions
```

## 四、Dify 集成外部知识库（RAGFlow）

### 1. 配置连接地址与认证信息

在 Dify 控制台中，进入「External Knowledge API」配置项：

- 填写 RAGFlow 的 API 地址（注意：若运行于 Docker 容器内，请使用 `host.docker.internal` 替代 `localhost`）；
- 添加 Authorization 请求头，填入 RAGFlow 生成的 API Key。

示例地址：
```
http://host.docker.internal:8080/api/v1/dify
```

### 2. 绑定知识库 ID

在 Dify 的应用配置中，明确指定目标知识库的唯一标识符（即 RAGFlow 中创建的知识库 ID），确保查询请求能精准路由至对应的数据源。

### 3. 测试集成效果

在 Dify 的聊天界面输入问题，观察是否能正确调用 RAGFlow 并返回结构化答案。可通过日志查看请求链路、响应时间与错误信息，确保集成稳定可靠。
